O objetivo desta atividade é criar um programa simples em Python para gerenciar um cadastro de compras e vendas de produtos.

Crie um menu interativo que permita ao usuário escolher entre as seguintes opções:

Registrar uma compra.

Registrar uma venda.

Verificar o saldo total das vendas.

Sair do programa.

Crie duas listas, uma para armazenar as compras e outra para armazenar as vendas. Cada registro nas listas deve conter informações como o nome do produto, a quantidade e o valor unitário.

Ao registrar uma compra, solicite ao usuário que insira o nome do produto, a quantidade e o valor unitário. Adicione essas informações à lista de compras.

Ao registrar uma venda, solicite ao usuário que insira o nome do produto, a quantidade e o valor unitário. Adicione essas informações à lista de vendas.

Ao verificar o saldo total das vendas, calcule o lucro total subtraindo o valor total das compras do valor total das vendas. Exiba o lucro na tela.

Certifique-se de que o programa possa lidar com erros, como entradas inválidas ou divisão por zero. Também atente-se a indentação ao enviar o código pelo sistema.

```python
# Lista com todos os elementos de cadastro de Compras
lista_compras = []

# Lista com todos os elementos de cadastro de Vendas
lista_vendas = []

#lista com o total faturado com todas as compras
compras_total = []

#lista com o total faturado com todas as vendas
vendas_total = []


# Um Loop infinito para inserir cadastro quantas as vezes quer for necessário 
while True:
        
        # Menu para Opção de Interesse
        print('\n',(30*'='))
        print(' Cadastrar  -  Compras / Vendas')
        print((30*'='), '\n')

        print("""
    Cadastrar Compra [ 1 ]        Cadastrar Venda   [ 2 ]

    Saldo de Vendas    [ 3 ]        Encerrar Programa [ 4 ]
      """)        

        # utilizando o comando "try" e "except" para lidar com qualquer erro
        try:
            
            # Digitando a opção de interesse no Menu
            menu = int(input('Selecione uma Opção do Menu: '))
            
            # Comando "if" para a opção de Cadastrar Compras
            if menu == 1:
                print('\n Cadastrar Compra \n')
                
                # Inserindo o Nome do Produto para compras
                nome = str(input('Nome do Produto: '))
                
                # Quantidade do Produto para compras
                qtd = int(input('Quantidade do Produto: '))
                
                # Valor R$ unitário do produto para compras
                unid_vl = float(input('Valor por Unidade R$ '))
                
                # Adicionando a Lista de Compras "Nome", "Quantidade" e "Valor por Unidade"
                lista_compras.append([nome, qtd, unid_vl])
                
                #Gerando valor total por item comprado
                compras_total.append(qtd * unid_vl)



             # Comando "if" para a opção de Cadastrar Vendas
            if menu == 2:
                print('\n  Cadastrar Venda  \n')
                
                 # Inserindo o Nome do Produto para vendas
                nome = str(input('Nome do Produto: '))
                
                # Quantidade do Produto para vendas
                qtd = int(input('Quantidade do Produto: '))
                
                # Valor R$ unitário do produto para vendas
                unid_vl = float(input('Valor por Unidade R$ '))
                
                 # Adicionando a Lista de Vendas "Nome", "Quantidade" e "Valor por Unidade"
                lista_vendas.append([nome, qtd, unid_vl])
                
                 #Gerando valor total por item vendido
                vendas_total.append(qtd * unid_vl)

            # comando "if" para opção do Menu "Saldo de Vendas e o Lucro"
            if menu == 3:
                print(f'\nSaldo Total em Vendas R$ {sum(vendas_total)}')
                print(f'Lucro Total Obtido em Vendas R$ {((sum(vendas_total)) - (sum(compras_total))):.2f}')

            # Comando "if" para a opção do Menu Encerrar Programa
            if menu == 4:
                print('\n Sessão Encerrada \n')
                break
                
        # Comando "except" para lidar com qualquer erro e não travar ou parar por erro no programa
        except:
            print('\nDigite Corretamente os Dados\n')

            

# Visualizando todos os Elemntos inseridos na lista de compras
print(f'\nCadastros Efetuados Compras {lista_compras}\n')

# Visualizando todos os Elemntos inseridos na lista de vendas
print(f'Cadasros Efetuados Vendas {lista_vendas}\n')

# Visualizando o Faturamento Total em Vendas
print(f'Saldo Total em Vendas R$ {sum(vendas_total):.2f}\n')

# Visualizando o Lucro do Faturamento descontando os gastos com as compras
print(f'Lucro Total Obtido em Vendas R$ {((sum(vendas_total)) - (sum(compras_total))):.2f}')
```
