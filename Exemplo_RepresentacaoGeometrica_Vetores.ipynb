### Exercício 1 - Representação Geométrica de Um Vetor

Neste exercício você deve representar geometricamente o vetor:

v = np.array([-1,2])

```python
# Vetor
v = np.array([-1,2])

# Dimensionando a figura para plotar
plt.figure(figsize=(6, 6))

# Criando  Flecha / Segmento de Reta que representa o vetor no grafico
vetor = plt.arrow(0, 0,  v[0], v[1], 
                 head_width = .3, width = .1, 
                  length_includes_head = True) 

# Criando "ponto Marcador" para a origem em amarelo 
plt.plot(0,0, 'ko', 
         markerfacecolor = 'yellow',  
         markersize = 8)

# Inserindo no plano uma Malha/Grid no Plot
plt.grid(linestyle = '--', linewidth = .6) 

plt.axis('square')
plt.axis([-3, 3, -3, 3])
plt.legend([vetor], ['v'])
plt.title(f'Vetor = {v}')
plt.savefig('Exercicio_Lista03.png', dpi = 300)
plt.show()
```

### Exercício 2 - Representação Geométrica da Soma dos Vetores

Neste exercício você deve representar geometricamente os 2 vetores abaixo e o vetor resultante da soma entre eles.

v = np.array([1,2])

w = np.array([4,-6])

```python
# Criando os Vetores
v = np.array([1,2])
w = np.array([4,-6])

# Craindo Variavel para a Soma
soma = v + w

print(f'1º Vetor: {v}\n2º Vetor: {w}')
# Imprimindo a Resultante da Soma de Vetores
print(f'A Resultante da Soma dos Vetores: {soma}')

# Criando as Setas / Flechas / Segmentos de reta para representar os Vetores
plt.figure(figsize=(6,6))
# 1º Vetor "v"
vetor_01 = plt.arrow(0, 0, v[0], v[1],
                    head_width = .3, width = .1, 
                     color='g', length_includes_head = True)

# 2º Vetor "w"
vetor_02 = plt.arrow(v[0], 
               v[1], 
               w[0], 
               w[1], 
               head_width = .3, 
               width = .1, 
               color = 'b', 
               length_includes_head = True)

# Vetor, Soma dos Vetores
v_soma = plt.arrow(0, 0, soma[0], soma[1],
                 head_width = .3, 
               width = .1, 
               color = 'r', 
               length_includes_head = True)



# Formata o plot
plt.grid(linestyle = '--', linewidth = .5)
plt.axis('square')
plt.axis([-8, 8, -8, 8])
plt.legend([vetor_01, vetor_02, v_soma],['v', 'w', 'v+w'])
plt.title('Vetores $\mathbf{v}$, $\mathbf{w}$ e $\mathbf{v+w}$')
plt.show()
```

### Exercício 3 - Representação Geométrica da Subtração de Vetores

Neste exercício você deve representar geometricamente os 2 vetores abaixo e o vetor resultante da subtração entre eles.

v = np.array([1,2])

w = np.array([4,-6])

```python
v01 = np.array([1,2])
w01 = np.array([4,-6])

subt = v01 - w01

print(f'1º Vetor: {v01}\n2º Vetor: {w01}')
# Imprimindo a Resultante da Soma de Vetores
print(f'A Resultante da Subtração dos Vetores: {subt}')

plt.figure(figsize=(6,6))

v_01 = plt.arrow(0, 0, v01[0], v01[1],
                 head_width = .3, width = .1, 
                 color='g', length_includes_head = True)

v_02 = plt.arrow(0,0, w01[0], w01[1], 
                 head_width = .3, width = .1, 
                 color='b', length_includes_head = True)

v_subt = plt.arrow(w[0],w[1], subt[0], subt[1],
                   head_width = .3, width = .1, 
                   color='r', length_includes_head = True)

# Inserindo Grade no Grafico
plt.grid(linestyle = '--', linewidth = .6) 

plt.axis('square')
plt.axis([-8, 8, -8, 8])
plt.legend([v_01, v_02, v_subt], ['v01', 'w01', 'subt'])
plt.title(f'Vetor = {v01}, {w01} e {subt}')
plt.savefig('Exercicio_03_lista03.png', dpi = 300)
plt.show()
```
